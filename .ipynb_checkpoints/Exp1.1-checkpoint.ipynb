Does the temperature vary within a position?

First, before I run this experiment, I want to flesh out the data 
1. make dummy data of shape num_gussets * x_coord * y_ coord,
2. populate with random data -5.00 < 0 < 20.00

```python
import numpy as np
n_gussets = 8
x_range = 3 #0,1,2
y_range = 30 #0....29

# Generate random dummy data between -5.00 and 20.00
dummy_data = np.random.uniform(-5.00, 20.00, (n_gussets, x_range, y_range))

# Print the shape of the data to verify
print(dummy_data.shape)
```

```python
import matplotlib.pyplot as plt

# Reshape the data to stack all gussets vertically for one plot
stacked_data = np.vstack([dummy_data[i] for i in range(n_gussets)])
print(stacked_data.shape)
# Create a single heatmap for all gussets stacked
plt.figure(figsize=(10, 8))
cax = plt.imshow(stacked_data, aspect='auto', cmap='coolwarm', vmin=-5, vmax=20)

# Add labels and title
plt.title('Temperature Variation Across Gussets')
plt.xlabel('X Coordinate')
plt.ylabel('Gussets and Y Coordinates')

# Add color bar
plt.colorbar(cax, label='Temperature (°C)')

# Mark individual gussets on the y-axis
gusset_separation = x_range * np.arange(1, n_gussets)  # y positions where gussets separate
print(gusset_separation)
for g in gusset_separation:
    plt.axhline(y=g-0.5, color='black', linestyle='--')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

n_gussets = 8
x_range = 3 #0,1,2
y_range = 30 #0....29

# Generate dummy data where temperature depends on the gusset number
# Let's say the base temperature for each gusset increases by 3 degrees per gusset
gusset_effect_dummy_data = np.array([np.random.uniform(-5.00, 20.00, (x_range, y_range)) + gusset_num * 3
                       for gusset_num in range(n_gussets)])

# Reshape the data to stack all gussets vertically for one plot
stacked_data = np.vstack([gusset_effect_dummy_data[i] for i in range(n_gussets)])
# Create a single heatmap for all gussets stacked
plt.figure(figsize=(10, 8))
cax = plt.imshow(stacked_data, aspect='auto', cmap='coolwarm', vmin=-5, vmax=20)

# Add labels and title
plt.title('Temperature Variation Across Gussets')
plt.xlabel('X Coordinate')
plt.ylabel('Gussets and Y Coordinates')
plt.colorbar(cax, label='Temperature (°C)')

# Mark individual gussets on the y-axis
gusset_separation = x_range * np.arange(1, n_gussets)  # y positions where gussets separate
for g in gusset_separation:
    plt.axhline(y=g-0.5, color='black', linestyle='--')

plt.tight_layout()
plt.show()
```

```python
n_gussets = 8
x_range = 3 #0,1,2
y_range = 30 #0....29

# Generate dummy data where temperature depends on the gusset number
# Let's say the base temperature for each gusset increases by 3 degrees per gusset
x_effect_dummy_data = np.array([
    np.random.uniform(-5.00, 20.00, (x_range, y_range)) + np.arange(x_range).reshape(-1, 1) * 3
    for gusset_num in range(n_gussets)
])

# Reshape the data to stack all gussets vertically for one plot
stacked_data = np.vstack([x_effect_dummy_data[i] for i in range(n_gussets)])
# Dynamically calculate the minimum and maximum values of the data
vmin = stacked_data.min()  # Minimum value in the data
vmax = stacked_data.max()  # Maximum value in the data

# Create a single heatmap for all gussets stacked horizontally
plt.figure(figsize=(10, 8))
cax = plt.imshow(stacked_data, aspect='auto', cmap='coolwarm', vmin=vmin, vmax=vmax)

# Add labels and title
plt.title('Temperature Variation Across X position within gusset')
plt.xlabel('X Coordinate')
plt.ylabel('Gussets and Y Coordinates')
plt.colorbar(cax, label='Temperature (°C)')

# Mark individual gussets on the y-axis
gusset_separation = x_range * np.arange(1, n_gussets)  # y positions where gussets separate
for g in gusset_separation:
    plt.axhline(y=g-0.5, color='black', linestyle='--')

plt.tight_layout()
plt.show()
```

```python
n_gussets = 8
x_range = 3 #0,1,2
y_range = 30 #0....29

# Generate dummy data where temperature depends on the gusset number
# Let's say the base temperature for each gusset increases by 3 degrees per gusset
y_effect_dummy_data = np.array([
    np.random.uniform(-5.00, 20.00, (x_range, y_range)) + np.arange(y_range).reshape(1,-1) * 3
    for gusset_num in range(n_gussets)
])

# Reshape the data to stack all gussets vertically for one plot
stacked_data = np.vstack([y_effect_dummy_data[i] for i in range(n_gussets)])
# Dynamically calculate the minimum and maximum values of the data
vmin = stacked_data.min()  # Minimum value in the data
vmax = stacked_data.max()  # Maximum value in the data

# Create a single heatmap for all gussets stacked horizontally
plt.figure(figsize=(10, 8))
cax = plt.imshow(stacked_data, aspect='auto', cmap='coolwarm', vmin=vmin, vmax=vmax)

# Add labels and title
plt.title('Temperature Variation Across Y position within gusset')
plt.xlabel('X Coordinate')
plt.ylabel('Gussets and Y Coordinates')
plt.colorbar(cax, label='Temperature (°C)')

# Mark individual gussets on the y-axis
gusset_separation = x_range * np.arange(1, n_gussets)  # y positions where gussets separate
for g in gusset_separation:
    plt.axhline(y=g-0.5, color='black', linestyle='--')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

# Assuming x_effect_dummy_data is already generated (n_gussets, x_range, y_range)
print("this is for the x effect data, where there hsould be variation along x but not y or gusset")
# 1. Test for variation across gussets
# We flatten x and y, and compare temperatures across gussets
flattened_across_gussets = x_effect_dummy_data.reshape(n_gussets, -1)
anova_across_gussets = stats.f_oneway(*flattened_across_gussets)

print("ANOVA across gussets:")
print(f"F-statistic: {anova_across_gussets.statistic}, p-value: {anova_across_gussets.pvalue}")

# 2. Test for variation across the x-axis (within gussets)
# We flatten y and compare x-coordinates within each gusset
flattened_across_x = x_effect_dummy_data.transpose(1, 0, 2).reshape(x_range, -1)
anova_across_x = stats.f_oneway(*flattened_across_x)

print("\nANOVA across x-coordinates (within gussets):")
print(f"F-statistic: {anova_across_x.statistic}, p-value: {anova_across_x.pvalue}")

# 3. Test for variation across the y-axis (within gussets)
# We flatten x and compare y-coordinates within each gusset
flattened_across_y = x_effect_dummy_data.transpose(2, 0, 1).reshape(y_range, -1)
anova_across_y = stats.f_oneway(*flattened_across_y)

print("\nANOVA across y-coordinates (within gussets):")
print(f"F-statistic: {anova_across_y.statistic}, p-value: {anova_across_y.pvalue}")
```
